```python
"""
AMMAD SOHAIL KARAR
TO_DO APPLICATION
1 MAKE list
2 ADD WHILE LOOP 
3 DISPLAY RESULT

"""
###to check whether the code number is right or not
def check(a):
    num_2=0
    for i in a:
        num_2+=1
    return num_2  
###to display final display
def final_display(b):
    print("------------------------")
    print("Things you will do today")
    for i in b:
        print("-->"+str(i))
###to display list display        
def display_list(b):
    num=0
    for i in b:
        print(str(num)+" : "+ i)
        num+=1
###to insert item into list    
def insert(a):
    insert=input("Enter what you want to do? ")
    a.append(insert)
###to delete item into list   
def delete(a):
    display_list(a)
    code=int(input("Enter the code number "))
    check_code=check(a)
    if code<=check_code:
        del a[code]
    else:
        print("Invalid Code")
    
def display():
    print("Welcome to ToDo list")
    print("Choice")
    print("Press I or i for Insert ")
    print("Press D or d for Delete")
    print("Press E or e for Exist")

to_do_list=[]
ch=""
display()
while ch!='e'and ch!='E':
    ch=input("Please write the choice ")
    
    if (ch=='i' or ch=='I'):
        insert(to_do_list)
    elif (ch=='d' or ch=='D'):
        delete(to_do_list)
    elif (ch!='e' and ch!='E'):
        print("Invalid Option Selected")
    display_list(to_do_list)    
final_display(to_do_list)
```
